# Scraping

## NASA Mars News

```python
#Get URL for NASA news
url = ('https://mars.nasa.gov/news/')

response = req.get(url)
soup = bs(response.text, 'html.parser')

print(soup.prettify())
```

```python
# pull titles from website
titles = soup.find_all('div', class_="content_title")
print(titles)
```

```python
# pull body from website
body = soup.find_all('div', class_="rollover_description")
print(body)
```

```python
# pull titles and body from website
results = soup.find_all('div', class_="slide")
for result in results:
    titles = result.find('div', class_="content_title")
    title = titles.find('a').text
    bodies = result.find('div', class_="rollover_description")
    body = bodies.find('div', class_="rollover_description_inner").text
    print('----------------')
    print(title)
    print(body)
```

## JPL Mars Space Images - Featured Image

```python
#Get URL for Space Images
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'

response = req.get(url)
soup = bs(response.text, 'html.parser')

print(soup.prettify())
```

```python
#Get Space Images
image_url = soup.find('article', class_='carousel_item')

footer = image_url.find('footer')
ref = footer.find('a')
path = ref['data-fancybox-href']

featured_image_url = ('https://www.jpl.nasa.gov' + path)
print(featured_image_url)
```

## Mars Weather

```python
#URL for twitter to get Mars Weather
url = 'https://twitter.com/marswxreport?lang=en'

response = req.get(url)
soup = bs(response.text, 'html.parser')

print(soup.prettify())
      
```

```python
contents = soup.find_all("div",class_="content")
```

```python
# Get weather details
weather_mars = []
for content in contents:
    tweet = content.find("div", class_="js-tweet-text-container").text
    weather_mars.append(tweet)
#print(weather_mars)

mars_weather = weather_mars[8]
print(mars_weather)
```
